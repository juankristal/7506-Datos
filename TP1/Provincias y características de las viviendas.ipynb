```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('Data/train.csv')
```

```python
viviendas = ["Casa", "Apartamento", "Casa en condominio", "Edificio", "Casa uso de suelo", "Quinta Vacacional", "Duplex", "Villa", "Departamento Compartido"]
df = df[df["tipodepropiedad"].isin(viviendas)]
```

```python
#Me quedo sólo con las propiedades que son viviendas
viviendas = ["Casa", "Apartamento", "Casa en condominio", "Edificio", "Casa uso de suelo", "Quinta Vacacional",
             "Duplex", "Villa", "Departamento Compartido"]
df = df[df["tipodepropiedad"].isin(viviendas)]
#Elimino las propiedades con un NAN en provincia
df = df.dropna(subset=["provincia"])
df = df.dropna(subset=["habitaciones"])
df = df.dropna(subset=["banos"])
df = df.fillna(0)
df = df[df["provincia"] != 0]
```

```python
promedios = df.groupby("provincia").agg({"habitaciones": "mean", "banos": "mean", "piscina": "mean", "garages": "mean", "gimnasio": "mean", "usosmultiples": "mean",})
promedios
```
