### Cuales son las 5 ciudades mas cara para comprar una propiedad?

```python
#Me quedo con las propiedades habitables
tipoDePropiedadHabitable = ['Casa', 'Apartamento', 'Casa en condominio', 'Duplex', 'Departamento Compartido', 'Lote']
propiedadesHabitables = df[df.tipodepropiedad.isin(tipoDePropiedadHabitable)]

#Elimino las propiedades que no tienen ciudad marcada
propiedadesHabitables = propiedadesHabitables.dropna(subset=['ciudad'])

#Elimino las propiedades ubicadas en ciudades con menos de 100 propiedades
ciudadCantPropiedades = propiedadesHabitables.ciudad.value_counts()
ciudadCantPropiedades = ciudadCantPropiedades[ciudadCantPropiedades[:] > 100]
propiedadesHabitables = propiedadesHabitables[propiedadesHabitables.ciudad.isin(ciudadCantPropiedades.index)]
```

```python
ciudadesMasCaras = propiedadesHabitables.groupby('ciudad')['precio'].mean()
ciudadesMasCaras = ciudadesMasCaras.to_frame()
ciudadesMasCaras.rename(columns = {'ciudad': 'Ciudad', 'precio': 'Precio promedio de propiedad'}, inplace=True)
ciudadesMasCaras.astype({'Precio promedio de propiedad': 'int32'})
ciudadesMasCaras.sort_values(by = 'Precio promedio de propiedad', ascending = False).head(5)
```

### Cuales son las 5 ciudades mas baratas para comprar una propiedad?
