Pruebo si estos datos mejoran los resultados de un xgboost

```python
#Transformo fecha a numero
df["fecha_n"] = df["fecha"].dt.year * 10000 + df["fecha"].dt.month * 100 + df["fecha"].dt.day
df["fecha_n"]

col_borrar = ["direccion", "lat", "lng", "titulo", "descripcion", "idzona", "fecha"]
df_1 = df.drop(col_borrar, axis=1)
```

```python
#Lidio con los nans, nada muy rebuscado

df_1.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer

df_1 = df_1.fillna(value = {'tipodepropiedad' : 'nan', 'provincia' : 'nan', 'ciudad' : 'nan'})

for c in ["antiguedad", "metrostotales", "metroscubiertos", "seguridad", "poblacion"]:
    imputer = SimpleImputer(strategy="mean")
    df_1[c] = imputer.fit_transform(df_1[[c]])

for c in ["habitaciones", "banos", "garages"]:
    imputer = SimpleImputer(strategy="constant", fill_value=0)
    df_1[c] = imputer.fit_transform(df_1[[c]])
    

```

```python
#Encodeo con un simple labelEncoder
from sklearn.preprocessing import LabelEncoder

codificador = LabelEncoder()
col_categoricas = ["provincia", "ciudad", "tipodepropiedad"]
col_datos = df_1.dtypes[df.dtypes != "object"].index
df_codificado = df_1[col_categoricas].apply(codificador.fit_transform)
df_2 = df_1[col_datos].join(df_codificado)
```

```python
df_2
```

```python
from sklearn.model_selection import train_test_split

datos = df_2.drop(["precio"], axis=1)
precios = df_2["precio"]

datos_train, datos_test, precio_train, precio_test = train_test_split(datos, precios, test_size=0.25, random_state=42)
```

```python
#XGBoost
from xgboost import XGBRegressor
```

```python
xgb_model = XGBRegressor(
    objective = 'reg:squarederror',
    subsample = 0.2,
    n_estimators = 7200,
    min_child_weight = 1.5,
    max_depth = 6,
    learning_rate = 0.05,
    reg_lambda = 0.6,
    gamma = 0.0,
    colsample_bytree = 0.2,
    seed = 42
)
```

```python
predictor = xgb_model.fit(datos_train, precio_train)
```

```python
prediccion = predictor.predict(datos_test)
```

```python
from common import metrica

metrica.resultados(precio_test, pd.Series(prediccion))
```
