```
X = train[FEATURES_1]
y = train[TARGET]
```

```
K = 5
def objective(hyper_parametros):

    base_estimator = RandomForestRegressor(
                          **hyper_parametros['random_forest']
                      )

    ada_regressor = AdaBoostRegressor(
                      base_estimator = base_estimator,
                      **hyper_parametros['adaboost']
                    )

    score = cross_val_score(ada_regressor, X, y, 
                            scoring='neg_mean_absolute_error', cv=K).mean()
    
    print("SCORE: {:.3f} params {}".format(score, hyper_parametros))
    
    return {'loss': -score, 'status': STATUS_OK}
```

```
space = {
    'random_forest' : {
                        'n_estimators': hp.randint("rf_n_estimators", 400) + 50,
                        'min_samples_split' : hp.randint('rf_min_samples_split', 1000) + 2,
                        'max_depth': hp.randint("rf_max_depth", 10) + 10,  
                        'max_features': hp.randint("rf_max_features", 10) + 10, 
                        'min_samples_leaf': hp.randint('rf_min_samples_leaf', 1000) + 1,
                      },
    'adaboost' :  {
                    'n_estimators': hp.randint("ada_n_estimators", 400) + 50,
                    'learning_rate' : hp.uniform('ada_learning_rate', 0.01, 1)
                  }
        }
```

```
N_ITER = 10
t0 = time.time()
best = fmin(fn=objective,
            space=space,
            algo=tpe.suggest,
            max_evals=N_ITER)
t1 = time.time()
print('Tiempo = {0:.2f} horas'.format((t1 - t0)/3600))
```
