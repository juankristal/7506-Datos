## Fill NaN: tipodepropiedad, antiguedad, habitaciones, garages, banos 

```python
out = mega_set.copy()
for col in  ['tipodepropiedad', 'idzona']:
    modas_x_ciudad_x_provincia =\
        mega_set.groupby(['provincia', 'ciudad'])[col]\
            .apply(lambda x: x.mode())\
            .reset_index()\
            .drop(['level_2'], axis = 1)\
            .rename(columns = {col : col + '_mode_x_ciudad'})\
            .set_index(['provincia', 'ciudad']).to_dict()
    
    out[col] = out.apply(lambda x: 
                         x[col] if not pd.isna(x[col]) 
                                else 
                                  modas_x_ciudad_x_provincia[col + '_mode_x_ciudad']\
                                    .get((x['provincia'], x['ciudad']), np.nan),
                        axis = 1)
    
    modas_x_provincia =\
        mega_set.groupby(['provincia'])[col]\
            .apply(lambda x: x.mode())\
            .reset_index()\
            .drop(['level_1'], axis = 1)\
            .rename(columns = {col : col + '_mode_x_provincia'})\
            .set_index(['provincia']).to_dict()
    
    out[col] = out.apply(lambda x: 
                         x[col] if not pd.isna(x[col])
                                else 
                                  modas_x_provincia[col + '_mode_x_provincia']\
                                    .get(x['provincia'], np.nan),
                        axis = 1)
    
    
    out.fillna(value = {col : out[col].mode().get(0)}, inplace = True)
mega_set = out.copy()
```

```python
out = mega_set.copy()
provincias_dict = pd.read_csv('/content/drive/My Drive/Colab Notebooks/data/ciudades_lat_lon.csv').to_dict()
ciudades_dict = pd.read_csv('/content/drive/My Drive/Colab Notebooks/data/provincias_lat_lon.csv').to_dict()

out["lat"] = out.apply(lambda x: x["lat"] if not np.isnan(x["lat"]) else ciudades_dict["Latitude"]\
                        .get(x["ciudad"],provincias_dict["Latitude"].get(x["provincia"], np.nan)), axis=1)

out["lng"] = out.apply(lambda x: x["lng"] if not np.isnan(x["lng"]) else ciudades_dict["Longitude"]\
                        .get(x["ciudad"],provincias_dict["Longitude"].get(x["provincia"], np.nan)), axis=1)

lat_lng_modas_x_idzona = mega_set.groupby(['idzona']).agg({'lat' : 'mean', 'lng' : 'mean'}).to_dict()

out['lat'] = out.apply(lambda x: 
                        x["lat"] if not pd.isna(x["lat"]) 
                                else lat_lng_modas_x_idzona["lat"]\
                                      .get(x["idzona"], np.nan), axis=1)

out['lng'] = out.apply(lambda x: 
                        x["lng"] if not pd.isna(x["lng"]) 
                                else lat_lng_modas_x_idzona["lng"]\
                                      .get(x["idzona"], np.nan), axis=1)

out.fillna(value = {'lat' : 23.062283, 'lng' : -109.699951}, inplace = True)
mega_set = out.copy()
```
