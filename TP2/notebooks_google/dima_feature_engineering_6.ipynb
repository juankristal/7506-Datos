```
import pandas as pd
import numpy as np
```

```
train = pd.read_csv("/content/drive/My Drive/Colab Notebooks/sets_de_datos/dima_train_with_features_5.csv")
```

```
X_test = pd.read_csv("/content/drive/My Drive/Colab Notebooks/sets_de_datos/dima_test_with_features_5.csv")
```

```
train['distancia_euclideana_al_origen'] = np.sqrt(np.square(train['lat']) + np.square(train['lng']))
train['distancia_minima_comercial'] = train.apply(lambda x: min(x['distancia_minima_Local_en_centro_comercial'], 
                                                                x['distancia_minima_Oficina_comercial'],
                                                                x['distancia_minima_Bodega_comercial']), axis = 1)
train['coseno_maximo_ciudad_pais'] = train.apply(lambda x: max(x[ 'coseno_distancia_angular_al_centro_pais'],
                                                               x['coseno_distancia_angular_al_centro_ciudad']),
                                                 axis = 1) 
train['ciudad_mean_antiguedad_sobre_provincia_mean_antiguedad'] =\
  train['ciudad_mean_antiguedad'] / (train['provincia_mean_antiguedad'] + 1)
train['tipodepropiead_mean_utilidades_extra_sobre_ciudad_mean_utilidades_extra'] =\
  train['tipodepropiedad_mean_utilidades_extra'] / (train['ciudad_mean_utilidades_extra'] + 1)
train['antiguedad_sobre_tipodepropiedad_mean_antiguedad'] = train['antiguedad'] / (train['tipodepropiedad_mean_antiguedad'] + 1)
```

```
train['direccion_cantidad_al_menos_una_mayuscula'] = train['direccion_cantidad_titulos'] + train['direccion_cantidad_palabras_en_mayuscula']
train['direccion_cantidad_fijos_top_k'] = train['direccion_cantidad_postfijos_top_k'] + train['direccion_cantidad_prefijos_top_k']
train['titulo_cantidad_fijos_top_k'] = train['titulo_cantidad_postfijos_top_k'] + train['titulo_cantidad_prefijos_top_k']
train['titulo_palabras_top_k_sobre_total_palabras'] =\
  train['titulo_cantidad_palabras_top_k'].astype(np.float64)\
  / (train['titulo_cantidad_palabras_importantes'].astype(np.float64) + train['titulo_cantidad_stopwords'].astype(np.float64) + 1)
train['ciudad_distancia_al_origen'] = np.sqrt(np.square(train['ciudad_mean_lng']) + np.square(train['ciudad_mean_lat']))
train['ciudad_mean_mean_todas'] = (train['ciudad_mean_garages']\
                                  + train['ciudad_mean_banos']\
                                  + train['ciudad_mean_utilidades_extra']\
                                  + train['ciudad_mean_habitaciones']\
                                  + train['ciudad_mean_servicios_cercanos']) / 5
train['ciudad_mean_antiguedad_sobre_mean_metrocubiertos'] = train['ciudad_mean_antiguedad'] / (train['ciudad_mean_metroscubiertos'] +1)
```
