```
train_rico = train[train['precio'] > (3390000.00 + 12525000.00) / 3]
```

```
train_pobre = train[train['precio'] <  (952772.50 + 310000.00) /2 ]
```

```
train_rico_lat_lng = train_rico[['lat', 'lng']].copy()
train_rico_lat_lng['lat_lng'] = train_rico_lat_lng.apply(lambda x: (x['lat'], x['lng']), axis = 1).copy()
train_rico_lat_lng = train_rico_lat_lng[['lat_lng']]
train_rico_lat_lng = train_rico_lat_lng.drop_duplicates()
train_rico_lat_lng['lat_lng'] = train_rico_lat_lng['lat_lng'].transform(lambda lista: [[x,y] for x,y in lista])
train_rico_lat_lng_dict = train_rico_lat_lng.to_dict()
```

```
train_pobre_lat_lng = train_pobre[['lat', 'lng']].copy()
train_pobre_lat_lng['lat_lng'] = train_pobre_lat_lng.apply(lambda x: (x['lat'], x['lng']), axis = 1).copy()
train_pobre_lat_lng = train_pobre_lat_lng[['lat_lng']]
train_pobre_lat_lng = train_pobre_lat_lng.drop_duplicates()
train_pobre_lat_lng['lat_lng'] = train_pobre_lat_lng['lat_lng'].transform(lambda lista: [[x,y] for x,y in lista])
train_pobre_lat_lng_dict = train_pobre_lat_lng.to_dict()
```

```
from sklearn.neighbors import NearestNeighbors
```

```
nbrs_ricos = NearestNeighbors(n_neighbors=10)
nbrs_ricos.fit(list(train_rico_lat_lng_dict['lat_lng'].values()))
nbrs_pobres = NearestNeighbors(n_neighbors=10)
nbrs_pobres.fit(list(train_pobre_lat_lng_dict['lat_lng'].values()))
```

```
train['distancia_minima_promedio_ricos'] =\
  train.apply(lambda x: 
                  np.mean(nbrs_ricos.kneighbors([[x['lat'], x['lng']]])[0]),
                  axis = 1)
train['distancia_minima_promedio_pobres'] =\
  train.apply(lambda x: 
                  np.mean(nbrs_pobres.kneighbors([[x['lat'], x['lng']]])[0]),
                  axis = 1)
```

```
X_test['distancia_minima_promedio_ricos'] =\
  X_test.apply(lambda x: 
                  np.mean(nbrs_ricos.kneighbors([[x['lat'], x['lng']]])[0]),
                  axis = 1)
X_test['distancia_minima_promedio_pobres'] =\
  X_test.apply(lambda x: 
                  np.mean(nbrs_pobres.kneighbors([[x['lat'], x['lng']]])[0]),
                  axis = 1)
```
