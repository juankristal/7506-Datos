```
mega_set = train.drop(['precio'], axis = 1).append(X_test, ignore_index = True, sort = False)
```

```
mega_set_lat_lng = mega_set[['tipodepropiedad', 'lat', 'lng']].copy()
```

```
mega_set_lat_lng['lat_lng'] = mega_set_lat_lng.apply(lambda x: (x['lat'], x['lng']), axis = 1)
```

```
mega_set_lat_lng_grupo = mega_set_lat_lng.groupby(['tipodepropiedad']).agg({'lat_lng' : 'unique'})
mega_set_lat_lng_grupo['lat_lng'] = mega_set_lat_lng_grupo['lat_lng'].transform(lambda lista: [[x,y] for x,y in lista])
lat_lng_todas = mega_set_lat_lng_grupo.to_dict()
```

```
from sklearn.neighbors import NearestNeighbors
```

```
todos_nbrs = {}
for tipodepropiedad in ['Apartamento', 'Casa en condominio', 'Casa', 'Terreno',
                        'Terreno comercial', 'Local Comercial', 'Quinta Vacacional',
                        'Oficina comercial', 'Edificio', 'Casa uso de suelo',
                        'Local en centro comercial', 'Bodega comercial', 'Otros', 'Villa',
                        'Duplex', 'Inmuebles productivos urbanos',
                        'Departamento Compartido', 'Nave industrial', 'Rancho',
                        'Terreno industrial', 'Huerta', 'Lote', 'Hospedaje', 'Garage']:
  nbrs = NearestNeighbors(n_neighbors=10)
  nbrs.fit(lat_lng_todas['lat_lng'][tipodepropiedad])
  todos_nbrs[tipodepropiedad] = nbrs
```

```
for tipodepropiedad in  ['Apartamento', 'Casa en condominio', 'Casa', 'Terreno',
                        'Terreno comercial', 'Local Comercial', 'Quinta Vacacional',
                        'Oficina comercial', 'Edificio', 'Casa uso de suelo',
                        'Local en centro comercial', 'Bodega comercial', 'Otros', 'Villa',
                        'Duplex', 'Inmuebles productivos urbanos',
                        'Departamento Compartido', 'Nave industrial', 'Rancho',
                        'Terreno industrial', 'Huerta', 'Lote', 'Hospedaje', 'Garage']:
      train['distancia_minima_' + "_".join(tipodepropiedad.split(' '))] =\
        train.apply(lambda x: 
                        np.mean(todos_nbrs[tipodepropiedad].kneighbors([[x['lat'], x['lng']]])[0]),
                        axis = 1)
```
