```python
from common import metrica
```

```python
df["fecha"] = pd.to_datetime(df["fecha"])
```

```python
from sklearn.model_selection import train_test_split
from category_encoders import BinaryEncoder
from sklearn.impute import SimpleImputer
from sklearn.tree import DecisionTreeRegressor
```

```python
col_borrar = ["direccion", "lat", "lng", "titulo", "descripcion", "idzona", "fecha"]
df = df.drop(col_borrar, axis=1)
df
```

```python
col_categorias = ["ciudad", "provincia", "tipodepropiedad"]
df[col_categorias].isnull().sum()
```

```python
# Si borro los nans me pierdo, a lo sumo (probablemente menos), 573 filas de 240000, para esta prueba simple no afecta tanto
df_categorico = df[col_categorias].dropna()
df_categorico
```

```python
col_datos = list(set(df.columns.tolist()) - set(col_categorias))
col_datos
```

```python
codificador = BinaryEncoder(cols=col_categorias)
df_binario = codificador.fit_transform(df_categorico)
df2 = df_binario.join(df[col_datos], how="inner")
df2
```

```python
df2.columns
```

```python
datos = df2.drop(["precio"], axis=1)
precios = df2["precio"]
datos_train, datos_test, precio_train, precio_test = train_test_split(datos, precios, test_size=0.25, random_state=1)
```

```python
col_a_imputar = list(set(df.columns[df.isnull().sum() > 0].tolist()) - set(col_categorias))
col_a_imputar
```

```python
for c in col_a_imputar:
    imp = SimpleImputer(strategy="median")
    datos_train[c] = imp.fit_transform(datos_train[[c]])
    datos_test[c] = imp.fit_transform(datos_test[[c]])
```

```python
display(datos_train.isnull().sum())
display(datos_test.isnull().sum())
```

```python
arbol = DecisionTreeRegressor().fit(datos_train, precio_train)
pred = arbol.predict(datos_test)
```

```python
metrica.resultados(precio_test, pd.Series(pred))
```
